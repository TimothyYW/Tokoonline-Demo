Aplikasi demo toko online

```python
class Produk:
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga

class TokoOnlineDemo:
    def __init__(self):
        self.produk = [
            Produk("Mouse", 75000),
            Produk("Keyboard", 105000),
            Produk("Flashdisk", 50000),
            Produk("Monitor", 15000),
            Produk("Charger", 85000)
        ]

    def selection_sort(self, key):
        n = len(self.produk)
        for i in range(n - 1):
            min_idx = i
            for j in range(i + 1, n):
                val_j = getattr(self.produk[j], key)
                val_min = getattr(self.produk[min_idx], key)
                if val_j < val_min:
                    min_idx = j
            self.produk[i], self.produk[min_idx] = self.produk[min_idx], self.produk[i]

    def tampilkan_produk(self):
        print("\nDaftar Produk:")
        for p in self.produk:
            print(f"- {p.nama}: Rp {p.harga}")

    def tambah_produk(self):
        nama = input("Masukkan nama produk: ")
        try:
            harga = int(input("Masukkan harga produk (angka): "))
            self.produk.append(Produk(nama, harga))
            print(f"Produk '{nama}' berhasil ditambahkan.")
        except ValueError:
            print("Harga harus berupa angka!")

    def jalankan(self):
        print("Selamat datang di toko online Demo")
        while True:
            print("\nMenu:")
            print("a. Urutkan berdasarkan harga")
            print("b. Urutkan berdasarkan nama")
            print("c. Tampilkan produk")
            print("d. Tambahkan produk baru")
            print("e. Keluar")

            option = input("Pilih opsi (a/b/c/d/e): ").lower()

            match option:
                case "a":
                    self.selection_sort("harga")
                    self.tampilkan_produk()
                case "b":
                    self.selection_sort("nama")
                    self.tampilkan_produk()
                case "c":
                    self.tampilkan_produk()
                case "d":
                    self.tambah_produk()
                case "e":
                    print("Terima kasih telah menggunakan aplikasi. Sampai jumpa!")
                    break
                case _:
                    print("Pilihan tidak valid. Silakan pilih a, b, c, d, atau e.")


if __name__ == "__main__":
    app = TokoOnlineDemo()
    app.jalankan()

```
